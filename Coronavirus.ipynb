```
import matplotlib.pyplot as plt
%matplotlib inline

uk_cases = [114, 160, 206, 271, 321, 373, 456, 590, 797, 1061, 1391, 1543, 1950, 2626, 3269, 3983]
uk_deaths = [1, 2, 2, 3, 5, 6, 8, 8, 10, 21, 35, 55, 71, 103, 144, 177]
uk_days = list(range(1, len(uk_cases) + 1))

italy_cases = [20, 79, 150, 229, 322, 400, 650, 888, 1128, 1694, 2036, 2502, 3089, 3858, 4636, 5883, 7375, 9172, 10149, 12462, 15113, 17660, 21157, 24747, 27980, 31506, 35713, 41035, 47021]
italy_deaths = [1, 2, 3, 6, 10, 12, 17, 21, 29, 34, 52, 79, 107, 148, 197, 233, 366, 463, 631, 827, 1016, 1266, 1441, 1809, 2158, 2503, 2978, 3405, 4032]
italy_days = list(range(1, len(italy_cases) + 1))

countries = [
    ("UK", uk_cases, uk_deaths, uk_days),
    ("Italy", italy_cases, italy_deaths, italy_days)
]
```

Case v death plot

```
for c in countries:
  plt.plot(c[1], label=f"{c[0]} Cases")
  plt.plot(c[2], label=f"{c[0]} Deaths")

plt.xlabel("Days since first death")
plt.ylabel("No. people")
plt.legend()

plt.show()
```

Case plot

```
for c in countries:
  plt.plot(c[1], label=f"{c[0]} Cases")

plt.xlabel("Days since first death")
plt.ylabel("No. people")
plt.legend()

plt.show()
```

Death plot

```
for c in countries:
  plt.plot(c[2], label=f"{c[0]} Deaths")

plt.xlabel("Days since first death")
plt.ylabel("No. people")
plt.legend()

plt.show()
```
